# Line Detection

```python
# Importing Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt  
```

```python
# function for displaying image
def imgShow(img):
    cv2.imshow("Image", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# Reading an image 
image = cv2.imread('lines.jpg')
imgShow(image)
```

```python
plt.imshow(image, cmap='gray')
```

```python
# Converting into grayscale image
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
imgShow(image_gray)
```

```python
plt.imshow(image_gray, cmap='gray')
```

## Canny Edge Detection

```python
image_canny = cv2.Canny(image_gray, 50, 220)
```

```python
image_canny
```

```python
plt.imshow(image_canny, cmap="gray")
```

```python
lines = cv2.HoughLines(image_canny, 1, np.pi/180, 100)
```

```python
lines
```

```python
len(lines)
```

```python
if lines is not None:
    for i in range(len(lines)):
        rho = lines[i][0][0]
        theta = lines[i][0][1]
        
        a = np.cos(theta)
        b = np.sin(theta)
        
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))

        cv2.line(image,(x1,y1),(x2,y2),(0,255,0),1)
```

```python
imgShow(image)
```
